# ***Developing the linear regression algorithm uisng Gradiant descent***

1_lIthvknHt9Tok5aIj4e__g.png

```python
import numpy as np 
import matplotlib.pyplot as plt
# define a sample dataset
n = 100
x = np.random.rand(n,1)
y = np.random.rand(1)*x + np.random.rand(n,1)*0.3
theta = np.random.rand(2,1)
print(theta.shape)
theta = theta.T
new_x=np.concatenate((np.ones((n,1)),x),axis=1)
%matplotlib inline
plt.ylabel("Labels")
plt.scatter(x,y, color="red", marker = "+")
y_pred=np.dot(theta,new_x.T)
print(x.shape)
print(y.shape)
# define the learning rate
alpha = 0.4
for i in range(1000):
  dev_y = -2/100*np.dot((y.T - y_pred),new_x)
  theta=theta-alpha*dev_y
  y_pred=np.dot(theta,new_x.T)
  
plt.plot(x,y_pred.T)
```

# **Mini Batch Gradient Descent**

```python
n = 100
x = np.random.rand(n,1)
y = np.random.rand(1)*x + np.random.rand(n,1)*0.3
theta = np.random.rand(2,1)
print(theta.shape)
theta = theta.T
new_x=np.concatenate((np.ones((n,1)),x),axis=1)
%matplotlib inline
plt.ylabel("Labels")
plt.scatter(x,y, color="red", marker = "+")
y_pred=np.dot(theta,new_x.T)
print(x.shape)
print(y.shape)
alpha = 0.4
epsilon=0.00001
b=10
#define a batch of size 10 
for i in range(0,len(x),b):
  dev_y = -2/10*np.dot((y[i:i+b].T - y_pred[0][i:i+b]),new_x[i:i+b])
  theta=theta-alpha*dev_y
  y_pred=np.dot(theta,new_x.T)

plt.plot(x,y_pred.T)
```

```python
n = 100
x = np.random.rand(n,1)
y = np.random.rand(1)*x + np.random.rand(n,1)*0.3
theta = np.random.rand(2,1)
print(theta.shape)
theta = theta.T
new_x=np.concatenate((np.ones((n,1)),x),axis=1)
%matplotlib inline
plt.ylabel("Labels")
plt.scatter(x,y, color="red", marker = "+")
y_pred=np.dot(theta,new_x.T)
print(x.shape)
print(y.shape)
alpha = 0.4
epsilon=0.00001
b=10
y=y.T
#At each iteration the value of theta is updated  
for i in range(len(x)):
  dev_y = -1*np.dot((y[0][i]- y_pred[0][i]),new_x[i])
  theta=theta-alpha*dev_y
  y_pred=np.dot(theta,new_x.T)

plt.plot(x,y_pred.T)
```
